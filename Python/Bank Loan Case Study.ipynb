```python
corr_non_deafulter = non_deafulter.corr()
corr_non_deafulter = corr_non_deafulter.where(np.triu(np.ones(corr_non_deafulter.shape),k=1).astype(bool)).unstack().reset_index()
corr_non_deafulter.columns =['VAR1','VAR2','Correlation']
corr_non_deafulter.dropna(subset = ["Correlation"], inplace = True)
corr_non_deafulter["Correlation"]=corr_non_deafulter["Correlation"].abs() 
corr_non_deafulter.sort_values(by='Correlation', ascending=False, inplace=True) 
corr_non_deafulter.head(10)
```
